```python
#1. Use Selenium Webdriver to get the URL 
from bs4 import BeautifulSoup
from selenium import webdriver
from datetime import datetime
import numpy as np
import pandas as pd
import requests
from datetime import timedelta 
import matplotlib.pyplot as plt
import time
%pip install html5lib


```

```python
driver = webdriver.Chrome()


#Store all of the data in a single DataFrame
all_data = pd.DataFrame()


#Note, need to run for all 8 pages
for page_number in range(1, 9):
    # Open the URL for the current page
    driver.get(f"https://amchainitiative.org/search-by-incident#incident/display-by-date/?view_279_per_page=1000&view_279_page={page_number}")
    
    # Wait for page to load and for it to find a table before progressing
    table = None
    while table is None:
        time.sleep(1)
        html = driver.page_source
        soup = BeautifulSoup(html, "html.parser")
        
        table = soup.find('table')

    
    df = pd.read_html(str(table))[0]  #Read the table into a DataFrame
    #Add this page's table to the full larger table
    all_data = pd.concat([all_data, df], ignore_index=True)
    
    

driver.quit()

```

```python
#Remove "incident details" column that just hyperlinks to view more
all_data = all_data.iloc[:, :-1]
```

```python
def print_percentage(data, keyword):
    filtered_data = data[data['Short Description'].str.contains(keyword, case=False)].copy()
    percentage = 100 * filtered_data.shape[0] / data.shape[0]
    print(f"Percentage of entries including {keyword}: {percentage:.2f}%")

# Usage
print_percentage(all_data, 'zionist|zionism')
print_percentage(all_data, 'jew|judaism')
print_percentage(all_data, 'Israel')
print_percentage(all_data, 'bds')
```

```python
# Get percentages based on content per year


def calculate_percentages(data, year):
    print(f"\n\n\n{year}")
    
    zionist = data[data['Short Description'].str.contains('zionist|zionism', case=False)].copy()
    print(f"Percentage of entries including Zionism or Zionist in {year}: {100 * round(zionist.shape[0] / data.shape[0], 4)}%")

    jew = data[data['Short Description'].str.contains('jew|judaism', case=False)].copy()
    print(f"Percentage of entries including Jew or Judaism in {year}: {100 * round(jew.shape[0] / data.shape[0], 4)}%")

    israel = data[data['Short Description'].str.contains('israel', case=False)].copy()
    print(f"Percentage of entries including Israel in {year}: {100 * round(israel.shape[0] / data.shape[0], 4)}%")

    bds = data[data['Short Description'].str.contains('bds', case=False)].copy()
    print(f"Percentage of entries including BDS in {year}: {100 * round(bds.shape[0] / data.shape[0], 4)}%")

    palestine = data[data['Short Description'].str.contains('palestine', case=False)].copy()
    print(f"Percentage of entries including Palestine in {year}: {100 * round(palestine.shape[0] / data.shape[0], 4)}%")

# Example usage
calculate_percentages(data_2015, 2015)
calculate_percentages(data_2016, 2016)
calculate_percentages(data_2017, 2017)
calculate_percentages(data_2018, 2018)
calculate_percentages(data_2019, 2019)
calculate_percentages(data_2020, 2020)
calculate_percentages(data_2021, 2021)
calculate_percentages(data_2022, 2022)
calculate_percentages(data_2023, 2023)
calculate_percentages(data_2024, 2024)




```
